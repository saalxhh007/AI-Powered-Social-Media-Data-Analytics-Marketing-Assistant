```python
import os
import re
import string
import json
import csv
    
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from nltk.corpus import stopwords, words as nltk_words
from nltk.stem import WordNetLemmatizer

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, confusion_matrix, accuracy_score, f1_score)
from sklearn.pipeline import Pipeline
from gensim.models import Word2Vec

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.svm import LinearSVC
from transformers import (AutoTokenizer, AutoModelForSequenceClassification,
                          TrainingArguments, Trainer)
from datasets import Dataset
import torch
from collections import Counter

import nltk
nltk.download('words')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
root_dir = "./data"

PLATFORM_MAP = {
    "twitter.com": "X",
    "instagram.com": "Instagram",
}
LABEL_MAP = {
    'BrandStory':    'Brand Story',
    'BTS':           'BTS',
    'Explainer':     'Explainer',
    'Humor':         'Humor',
    'IndustryNews':  'Industry News',
    'Inspirational': 'Inspirational',
    'LifestyleDaily':'Lifestyle Daily',
    'ProductLaunch': 'Product Launch',
    'SalesPromotion':'Sales Promotion',
    'TipsAndTricks': 'Tips And Tricks',
    'Tutorial':      'Tutorial',
}

def get_platform(record: dict) -> str:
    """Extract platform label from the source_platform field."""
    source = record.get("source_platform", "")
    for domain, label in PLATFORM_MAP.items():
        if domain in source:
            return label
    return "Unknown"

def get_text(record: dict, platform: str) -> str:
    data = record.get("data", {})

    if platform == "X":
        try:
            return data["note_tweet"]["note_tweet_results"]["result"]["text"]
        except (KeyError, TypeError):
            pass
        try:
            return data["legacy"]["full_text"]
        except (KeyError, TypeError):
            pass

    elif platform == "Instagram":
        try:
            return data["caption"]["text"]
        except (KeyError, TypeError):
            pass

    return record.get("text", "")

def process_ndjson(input_path: str, label: str, output_path: str):
    rows = []
    with open(input_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            record = json.loads(line)
            platform = get_platform(record)
            text = get_text(record, platform)
            if text:
                rows.append({"text": text, "label": label, "platform": platform})

    with open(output_path, "w", encoding="utf-8", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=["text", "label", "platform"])
        writer.writeheader()
        writer.writerows(rows)

    print(f"✓ {output_path}  ({len(rows)} rows)")

def main():
    for category in os.listdir(root_dir):
        category_path = os.path.join(root_dir, category)
        if not os.path.isdir(category_path):
            continue

        for filename in os.listdir(category_path):
            if not filename.endswith(".ndjson"):
                continue

            input_path = os.path.join(category_path, filename)
            output_filename = filename.replace(".ndjson", ".csv")
            output_path = os.path.join(category_path, output_filename)

            mapped_label = LABEL_MAP.get(category, category)
            process_ndjson(input_path, label=mapped_label, output_path=output_path)
            
if __name__ == "__main__":
    main()
```

```python
english_vocab = set(w.lower() for w in nltk_words.words())
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean(text):
    if not isinstance(text, str): return ""
    text = text.lower()
    # Remove URLs 
    text = re.sub(r'http\S+|www\S+|@\w+|@\w+|#', '', text)
    text = re.sub(r"@\w+|#\w+", "", text) 
    # Punctuation and Numbers
    text = text.translate(str.maketrans('', '', string.punctuation))
    # text = re.sub(r'\d+', '', text)
    
    stop_words = set(stopwords.words('english'))
    tokens = text.split()
    
    clean_tokens = [w for w in tokens if w not in stop_words and len(w) > 2]
    return " ".join(clean_tokens)

main_df = pd.read_csv('./data/master_training_dataset.csv')
```

```python
le = LabelEncoder()
main_df["label_enc"] = le.fit_transform(main_df["label"])
CLASSES = le.classes_

X_train, X_test, y_train, y_test = train_test_split(
    main_df["text"], main_df["label_enc"],
    test_size=0.2, random_state=42, stratify=main_df["label_enc"]
)

print(f"Train: {len(X_train)} | Test: {len(X_test)}")

X_train_c = X_train.map(clean)
X_test_c  = X_test.map(clean)
```

```python
res = {}

def evaluate(name, y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    f1  = f1_score(y_true, y_pred, average="weighted")
    rep = classification_report(y_true, y_pred, target_names=CLASSES)
    cm  = confusion_matrix(y_true, y_pred)
    res[name] = dict(accuracy=acc, f1=f1, report=rep, cm=cm)
    print(f"\n{'='*55}")
    print(f"  {name}")
    print(f"{'='*55}")
    print(f"  Accuracy : {acc:.4f}   Weighted F1 : {f1:.4f}")
    print(rep)

def plot_cm(name, cm):
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=CLASSES, yticklabels=CLASSES)
    plt.title(f"Confusion Matrix — {name}", fontsize=14)
    plt.ylabel("True Label"); plt.xlabel("Predicted Label")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    # plt.savefig(f"./plots/cm_{name.replace(' ','_')}.png", dpi=150)
    plt.show()
```

```python
# SVM
# Bag Of Words
print("\nRunning: BoW + SVM")
bow_pipe = Pipeline([
    ("vec", CountVectorizer(max_features=50_000, ngram_range=(1, 2))),
    ("clf", LinearSVC(max_iter=2000, C=1.0))
])
bow_pipe.fit(X_train_c, y_train)
y_pred_bow = bow_pipe.predict(X_test_c)
evaluate("BoW + SVM", y_test, y_pred_bow)
plot_cm("BoW + SVM", res["BoW + SVM"]["cm"])

# TF-IDF
print("\nRunning: TF-IDF + SVM")
tfidf_pipe = Pipeline([
    ("vec", TfidfVectorizer(max_features=50_000, ngram_range=(1, 2),
                            sublinear_tf=True)),
    ("clf", LinearSVC(max_iter=2000, C=1.0))
])
tfidf_pipe.fit(X_train_c, y_train)
y_pred_tfidf = tfidf_pipe.predict(X_test_c)
evaluate("TF-IDF + SVM", y_test, y_pred_tfidf)
plot_cm("TF-IDF + SVM", res["TF-IDF + SVM"]["cm"])

# WORD2VEC
print("\n>>> Running: Word2Vec + SVM")

def avg_w2v(texts, model, size):
    vecs = []
    for t in texts:
        tokens = t.split()
        v = [model.wv[w] for w in tokens if w in model.wv]
        vecs.append(np.mean(v, axis=0) if v else np.zeros(size))
    return np.array(vecs)

W2V_DIM = 200
tokenized_train = [t.split() for t in X_train_c]
w2v_model = Word2Vec(tokenized_train, vector_size=W2V_DIM,
                     window=5, min_count=2, workers=4, epochs=10)

X_train_w2v = avg_w2v(X_train_c, w2v_model, W2V_DIM)
X_test_w2v  = avg_w2v(X_test_c,  w2v_model, W2V_DIM)

w2v_svm = LinearSVC(max_iter=3000, C=1.0)
w2v_svm.fit(X_train_w2v, y_train)
y_pred_w2v = w2v_svm.predict(X_test_w2v)
evaluate("Word2Vec + SVM", y_test, y_pred_w2v)
plot_cm("Word2Vec + SVM", res["Word2Vec + SVM"]["cm"])
```

```python
MODEL_PATH = "./bertweet_topic_detection_model" 

# 2. Load the objects
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)
model_bert = AutoModelForSequenceClassification.from_pretrained(MODEL_PATH)

# 3. Move to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_bert.to(device)
model_bert.eval() # Set to evaluation mode

def softmax(x):
    e = np.exp(x - np.max(x))
    return e / e.sum()

def predict_all_models(text: str):
    text_clean = clean(text)
    predictions = {}

    # --- BoW + SVM ---
    pred  = bow_pipe.predict([text_clean])[0]
    score = softmax(bow_pipe.decision_function([text_clean])[0])
    predictions["BoW + SVM"] = (le.inverse_transform([pred])[0], score.max())

    # --- TF-IDF + SVM ---
    pred  = tfidf_pipe.predict([text_clean])[0]
    score = softmax(tfidf_pipe.decision_function([text_clean])[0])
    predictions["TF-IDF + SVM"] = (le.inverse_transform([pred])[0], score.max())

    # --- Word2Vec + SVM ---
    vec   = avg_w2v([text_clean], w2v_model, W2V_DIM)
    pred  = w2v_svm.predict(vec)[0]
    score = softmax(w2v_svm.decision_function(vec)[0])
    predictions["Word2Vec + SVM"] = (le.inverse_transform([pred])[0], score.max())

    # --- BERTweet ---
    model_bert.eval()
    # No need to redefine device if done globally, but safe to check:
    current_device = next(model_bert.parameters()).device
    
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding="max_length",
        max_length=128
    ).to(current_device)

    with torch.no_grad():
        logits = model_bert(**inputs).logits
    
    probs = torch.softmax(logits, dim=-1).squeeze().cpu().numpy()
    pred_int = probs.argmax()
    
    # Map the integer prediction to the class name
    # Ensure 'le' or a class list is defined!
    predicted_label = le.inverse_transform([pred_int])[0]
    predictions["BERTweet"] = (predicted_label, probs.max())

    # ---- Print table ----
    print("\n" + "="*62)
    print(f'  INPUT: "{text[:70]}{"..." if len(text)>70 else ""}"')
    print("="*62)
    print(f"  {'Model':<20} | {'Predicted Class':<20} | Confidence")
    print("-"*62)
    for model_name, (label, conf) in predictions.items():
        print(f"  {model_name:<20} | {label:<20} | {conf:.2%}")

    # Majority vote
    majority = Counter(v[0] for v in predictions.values()).most_common(1)[0][0]
    print("-"*62)
    print(f"  {'Majority Vote':<20} | {majority:<20} |")
    print("="*62)

    # ---- Plot ----
    plot_prediction_graph(text, predictions)
    return predictions


def plot_prediction_graph(text: str, predictions: dict):
    """
    Network graph where:
    - Center node  = the input text (gray)
    - Outer nodes  = predicted classes (one per model)
    - Edges        = model name + confidence
    - Node color   = class (same class = same color across models)
    - Node size    = confidence
    """
    import networkx as nx
    from matplotlib.colors import Normalize
    from matplotlib.patches import Patch

    G = nx.DiGraph()
    CENTER = "INPUT"
    G.add_node(CENTER)

    model_names  = list(predictions.keys())
    pred_labels  = [predictions[m][0] for m in model_names]
    pred_confs   = [predictions[m][1] for m in model_names]

    # Assign a color per unique predicted class
    unique_preds = list(set(pred_labels))
    palette      = plt.cm.Set2(np.linspace(0, 1, max(len(CLASSES), 1)))
    class_color  = {cls: palette[i % len(palette)] for i, cls in enumerate(CLASSES)}

    # Add model nodes and edges from center
    for m, label, conf in zip(model_names, pred_labels, pred_confs):
        node_id = f"{m}\n→ {label}"
        G.add_node(node_id, label=label, conf=conf, model=m)
        G.add_edge(CENTER, node_id, model=m, conf=conf)

    # Layout: center node in the middle, models around it
    pos = {CENTER: (0, 0)}
    n   = len(model_names)
    for i, m in enumerate(model_names):
        node_id    = f"{m}\n→ {predictions[m][0]}"
        angle      = 2 * np.pi * i / n
        pos[node_id] = (1.6 * np.cos(angle), 1.6 * np.sin(angle))

    fig, ax = plt.subplots(figsize=(12, 10))
    ax.set_facecolor("#f4f4f8")
    fig.patch.set_facecolor("#f4f4f8")

    # Draw center node
    nx.draw_networkx_nodes(
        G, pos, nodelist=[CENTER],
        node_size=4000, node_color="#555555",
        linewidths=2, edgecolors="white", ax=ax
    )
    nx.draw_networkx_labels(
        G, pos, labels={CENTER: "INPUT\nTEXT"},
        font_size=9, font_color="white", font_weight="bold", ax=ax
    )

    # Draw model nodes — sized by confidence, colored by predicted class
    for m, label, conf in zip(model_names, pred_labels, pred_confs):
        node_id = f"{m}\n→ {label}"
        nx.draw_networkx_nodes(
            G, pos, nodelist=[node_id],
            node_size=int(conf * 8000),
            node_color=[class_color[label]],
            linewidths=2, edgecolors="white", ax=ax
        )

    # Node labels for model nodes
    model_node_labels = {
        f"{m}\n→ {predictions[m][0]}": f"{m}\n{predictions[m][0]}\n{predictions[m][1]:.1%}"
        for m in model_names
    }
    nx.draw_networkx_labels(
        G, pos, labels=model_node_labels,
        font_size=8, font_weight="bold", font_color="#222222", ax=ax
    )

    # Draw edges with confidence as width
    for m, label, conf in zip(model_names, pred_labels, pred_confs):
        node_id = f"{m}\n→ {label}"
        nx.draw_networkx_edges(
            G, pos, edgelist=[(CENTER, node_id)],
            width=1 + conf * 5,
            alpha=0.5 + conf * 0.4,
            edge_color=[class_color[label]],
            arrows=True,
            arrowstyle="-|>",
            arrowsize=20,
            connectionstyle="arc3,rad=0.05",
            ax=ax
        )

    # Edge labels = confidence %
    edge_labels = {
        (CENTER, f"{m}\n→ {predictions[m][0]}"): f"{predictions[m][1]:.1%}"
        for m in model_names
    }
    nx.draw_networkx_edge_labels(
        G, pos, edge_labels=edge_labels,
        font_size=8, font_color="#333333",
        bbox=dict(boxstyle="round,pad=0.2", fc="white", alpha=0.7),
        ax=ax
    )

    # Legend: unique predicted classes
    legend_patches = [
        Patch(color=class_color[c], label=c) for c in unique_preds
    ]
    ax.legend(handles=legend_patches, title="Predicted Class",
              loc="lower right", fontsize=9, title_fontsize=10)

    # Majority vote annotation
    majority = Counter(pred_labels).most_common(1)[0][0]
    ax.text(0, -2.3,
            f"🗳️  Majority Vote: {majority}",
            ha="center", fontsize=12, fontweight="bold",
            color="#222222",
            bbox=dict(boxstyle="round,pad=0.4", fc="white", alpha=0.8))

    title_text = text[:80] + ("..." if len(text) > 80 else "")
    ax.set_title(f'"{title_text}"\n'
                 f'Node size = confidence | Color = predicted class',
                 fontsize=11, fontweight="bold", pad=15)
    ax.axis("off")
    plt.tight_layout()
    plt.savefig("prediction_graph.png", dpi=150, bbox_inches="tight")
    plt.show()


# ============================================================
#  RUN IT
# ============================================================

# Single text
predict_all_models(
    "Just dropped our new skincare line 🧴✨ Get 20% off this weekend only! Link in bio 👇"
)

# Or loop over several
test_texts = [
    "Step-by-step guide to setting up your first Python environment 🐍",
    "Can't stop laughing at this video of a cat stealing pizza 😂😂",
    "Excited to announce we're launching our new app next Monday! 🚀",
    "Behind the scenes with our team getting ready for the big shoot 📸",
    "The Fed just raised interest rates again — what this means for crypto",
    "Morning routine that changed my life ☀️ #lifestyle #selfcare",
    "Success is not final, failure is not fatal. Keep going 💪 #motivation",
]

for t in test_texts:
    predict_all_models(t)
```
